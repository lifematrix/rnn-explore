```python
import numpy as np
import skimage.io
import matplotlib.pylab as plt
```

```python
# Inspired by <https://datascience.stackexchange.com/questions/30602/how-to-implement-pca-color-augmentation-as-discussed-in-alexnet>
# Covariance matrix using tensorflow <https://stackoverflow.com/questions/47709854/how-to-get-covariance-matrix-in-tensorflow?rq=1>
# Eigen Value and Vectors <https://www.tensorflow.org/api_docs/python/tf/self_adjoint_eig>

# # original image is 224x224x3 of dtype uint8 

# renorm_image = np.reshape(original_image,(original_image.shape[0]*original_image.shape[1],3))

# renorm_image = renorm_image.astype('float32')
# renorm_image -= np.mean(renorm_image, axis=0)
# renorm_image /= np.std(renorm_image, axis=0)

# cov = np.cov(renorm_image, rowvar=False)

# lambdas, p = np.linalg.eig(cov)
# alphas = np.random.normal(0, 0.1, 3)

# delta = np.dot(p, alphas*lambdas)

# delta = (delta*255.).astype('int8')

# pca_color_image = np.maximum(np.minimum(original_image + delta, 255), 0).astype('uint8')

def pca_color_aug(img, threshold=0.1, seed=None, normalize=True):
    if len(img.shape) != 3 and img.shape[2] != 3:
        return None
    r_img = img.reshape(-1, 3)
    r_img = r_img.astype(np.float32)
    if normalize:
        mean, std = np.mean(r_img, axis=0), np.std(r_img, axis=0)
        if np.any(std == 0.0):
            return None
        r_img = (r_img - mean)/std
    
    # co-variance matrix
    cov = np.cov(r_img, rowvar=False)
    # eigen values and vectors of cov matrix
    vals, vecs = np.linalg.eig(cov)
    
    if seed is not None:
        np.random.seed(seed)
    alphas = np.random.normal(0, threshold, 3)
    delta = np.dot(vecs, alphas*vals)
    r_img += delta
    
    if normalize:
        r_img = r_img * std + mean
    r_img = np.maximum(np.minimum(r_img, 255), 0)
    r_img = r_img.astype(np.uint8)
    r_img = r_img.reshape(img.shape)
    
    return r_img
```

```python
fname = "/Users/liucongxin/Documents/Airdoc/data/food/food-101/images/apple_pie/134.jpg"
img = skimage.io.imread(fname)
print(np.mean(img))
plt.imshow(img)
plt.show()
%time r_img = pca_color_aug(img, threshold=0.5, seed=1234, normalize=True)
print(np.mean(r_img))
plt.imshow(r_img)
plt.show()
```
